```python
NSIDE=256
NPIX = hp.nside2npix(NSIDE)
print(NPIX)
```

```python
m = np.arange(NPIX)
hp.mollview(m, title="Mollview image RING")
hp.graticule()

```

```python

vec = hp.ang2vec(np.pi/2, 0.)

fsky = 0.4

radius = fsky* np.radians(180)

mask = healpy.query_disc(NSIDE, vec = vec, radius=radius)

```

```python
m = np.zeros(NPIX)
m[mask] = 1
hp.mollview(m, title="Mollview image RING")

hp.write_map("disk_fsky_04.fits", m, overwrite=True)
```
